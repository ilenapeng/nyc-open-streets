```python
import pandas as pd
```

```python
df = pd.read_csv('Open_Streets_Locations.csv')
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
# add a space in staten island
df['borough_name'] = df.borough_name.str.replace('StatenIsland', 'Staten Island')
```

```python
df_count = df.groupby('borough_name').size().sort_values().reset_index(name='count')
```

```python
df_pop = pd.read_csv('nyc-boroughs.csv')
df_pop.head()
```

```python
# merge df
df_merge = df_count.merge(df_pop, left_on='borough_name', right_on='borough')
df_merge
```

```python
df_merge['per_capita'] = df_merge['count'] / df_merge['population']
df_merge
```

```python
df_merge['streets_per_person'] = df_merge['per_capita'] * 1000000
df_merge
```

```python
df_merge.to_csv('borough-count.csv', index=False)
```
